```python
# python、的2d绘图工具
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [0, 1]
y = [0, 1]
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, num=100)
```

```python
y = np.sin(x)
cosy = np.cos(x)
```

```python
plt.plot(x, y)
plt.plot(x, cosy)
plt.show()
```

# 散点图

```python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha =0.1)
plt.show()
```
